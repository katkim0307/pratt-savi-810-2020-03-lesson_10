# Pizza Analogy

![pizza](https://media.timeout.com/images/105294390/630/472/image.jpg)

When preparing and making some food or some dish you need to divide the tasks into a  hierarchy of tasks and then split those tasks up into the individual steps. 

So below we've outlined some of the steps necessary to make pizza at home. 


## 1. Get Pizza Making Ingredients
* Go to store
* Get dough off shelf
    * Place dough in cart
* Get sauce off shelf
    * Place sauce in cart
* Get cheese off shelf
    * Place cheese in cart (repetitive, 🤔 - could I use some Python trick here?)
* Checkout at register
* Carry ingredients in grocery bag to kitchen
    


## 2. Prepare Pizza Making Ingredients and Supplies
* For dough
    * Remove dough from grocery bag. 
* For sauce
    * Remove Sauce from grocery bag. 
* For cheese
    * Remove cheese from grocery bag. (repetitive, 🤔 - could I use some Python trick here?)
* For item in `[dough, sauce, cheese]`: 
    * Place item on kitchen counter
* For item in `[dough, sauce, cheese]`: 
    * Open item from packaging
* For item in `[dough, sauce, cheese]`: 
    * Place item on pizza sheet

## 3. Pizza Cooking Steps
* Set oven to 450° F
* While oven < 450° F
    * Wait, when True, pass to next step
* While cheese is not melted:
    * Wait, when True, pass to next step
* When cheese is melted:
    * Turn oven off


## 4. Present Pizza to table 
* Place oven mits on hands
* Carefully remove pizza from oven
* While pizza too hot to serve:
    * Wait
* When pizza at serving temperature
    * Slice with pizza cutter
* Carry pizza tray to dining table


## Creating High Level Task Functions

For a refresh, here's how a function is structured. 

```python

def a_function(argument_1, argument_2):
    new_variable = do_something(argument_1, argument_2)
    return new_variable 

```

Within a function, you can also have loops inside of a function. 

```python

def a_function(argument_1_list, argument_2):
    result_list = []
    
    for list_item in argument_1_list:
        result_list.append(
            do_something(argument_1, argument_2)
        )
    
    return result_list

```


Or have a loop occur loops outside of a function. 

```python

def a_function(argument_1, argument_2):
    return do_something(argument_1, argument_2)
    

result_list = []

for item in list_of_items:
    result_list.append(
        a_function(item, argument_2)
    )

result_list

```

```python
from travel import walk
from store_actions import grab_off_shelf, place_on_register, checkout


def get_ingredients(home, store, items_list): 
    walk(home, store)
    
    items_in_cart = []
    for item in items_list:
        
        item_in_hand = grab_off_shelf(item)  # Get dough off shelf
        
        items_in_cart.append(
            item_in_hand
        )  # Place item in cart

    bag_items = []
    for item in items_in_cart:  
        
        place_on_register(item)
        
        checkout(item)  # Checkout at register
        
        bag_items.append(item)
    
    walk(store, home)  # * Carry ingredients in grocery bag to kitchen
                
    return bag_items
```
